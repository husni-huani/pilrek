```python
import pandas as pd
```

```python
csv_a = '../pilrek-anon.csv'
df = pd.read_csv(csv_a)
```

```python
df.head()
```

```python
df['CaRek Pilihan'].value_counts()
```

```python
# DATAFRAME 10 BAKAL CALON REKTOR #
# =============================== #

# Benyamin Sapiie (FITB)
Benyamin_Sapiie_df = df.loc[df['CaRek Pilihan'] == 'Benyamin Sapiie (FITB)']


# Bramantyo (Direktur PPM Manajemen)
Bramantyo_df = df.loc[df['CaRek Pilihan'] == 'Bramantyo (Direktur PPM Manajemen)']

# Dwi Larso (SBM / President Univ)
Dwi_Larso_df = df.loc[df['CaRek Pilihan'] == 'Dwi Larso (SBM / President Univ)']


# Edy Tri Baskoro (FMIPA-MA)
Edy_Tri_df = df.loc[df['CaRek Pilihan'] == 'Edy Tri Baskoro (FMIPA-MA)']

# Gusti Ayu Putri Saptawati S (STEI)
Gusti_Ayu_df = df.loc[df['CaRek Pilihan'] == 'Gusti Ayu Putri Saptawati S (STEI)']

# Jaka Sembiring (STEI)
Jaka_Sembiring_df = df.loc[df['CaRek Pilihan'] == 'Jaka Sembiring (STEI)']

# Kadarsah Suryadi (FTI-TI)
Kadarsah_Suryadi_df = df.loc[df['CaRek Pilihan'] == 'Kadarsah Suryadi (FTI-TI)']

# Reini D Wirahadikusumah (FTSL)
Reini_D_df = df.loc[df['CaRek Pilihan'] == 'Reini D Wirahadikusumah (FTSL)']

# Togar M Simatupang (SBM)
Togar_M_df = df.loc[df['CaRek Pilihan'] == 'Togar M Simatupang (SBM)']

# Widjaja Martokusumo (SAPPK)
Widjaja_Martokusumo_df = df.loc[df['CaRek Pilihan'] == 'Widjaja Martokusumo (SAPPK)']



# meringkas variabel df
df1 = Benyamin_Sapiie_df
df2 = Bramantyo_df
df3 = Dwi_Larso_df
df4 = Edy_Tri_df
df5 = Gusti_Ayu_df
df6 = Jaka_Sembiring_df
df7 = Kadarsah_Suryadi_df
df8 = Reini_D_df
df9 = Togar_M_df
df10 = Widjaja_Martokusumo_df
```

```python
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10]
df_10R = pd.concat(frames)
df_10R
```
