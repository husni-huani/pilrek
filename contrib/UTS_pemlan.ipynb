```python
alasan_series = data['Alasan Memilih CaRek'].value_counts()
alasan_series_2 = data2['Alasan Memilih CaRek'].value_counts()
```

```python
alasan_other = []
for alasan in alasan_series.index[:12]:
    alasan_other.append(alasan)
alasan_other.append('Other')

alasan_other_2 = []
for alasan_2 in alasan_series_2.index[:12]:
    alasan_other_2.append(alasan_2)
alasan_other_2.append('Other')
```

```python
alasan_other
```

```python
jumlah_alasan = []
for i in alasan_series[:12]:
    jumlah_alasan.append(i)
count_other = 0
for i in alasan_series[12:]:
    count_other += i
jumlah_alasan.append(count_other)

jumlah_alasan_2 = []
for i in alasan_series_2[:12]:
    jumlah_alasan_2.append(i)
count_other_2 = 0
for i in alasan_series_2[12:]:
    count_other_2 += i
jumlah_alasan_2.append(count_other_2)
    
```

```python
sum = 0
for i in jumlah_alasan:
    sum += i
for i in range(len(jumlah_alasan)):
    jumlah_alasan[i] = jumlah_alasan[i]/sum

sum_2 = 0
for i in jumlah_alasan_2:
    sum_2 += i
for i in range(len(jumlah_alasan_2)):
    jumlah_alasan_2[i] = jumlah_alasan_2[i]/sum_2
```

```python
len(jumlah_alasan)
```

```python
alasan_other_series = pd.Series(jumlah_alasan, index=alasan_other)
alasan_other_series_2 = pd.Series(jumlah_alasan_2, index=alasan_other_2)
```

```python
alasan_other_series
```

```python
alasan_other_series.plot(kind= 'bar')
```

```python
alasan_gabung = pd.concat([alasan_other_series, alasan_other_series_2], axis = 1, keys=['Rektor_terpilih', 'Rektor_tidak_terpilih'])
```

```python
alasan_gabung = alasan_gabung.sort_values(by='Rektor_terpilih', ascending=False)
```

```python
alasan_gabung.plot(kind='bar', y=['Rektor_terpilih', 'Rektor_tidak_terpilih'])
```
